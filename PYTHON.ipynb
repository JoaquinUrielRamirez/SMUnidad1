### Genera una lista de 100 elementos con múltiplos de 3. E.g. [3, 6, 9, ...]

```python
lista = []
```

```python
elelista = len(lista)
```

```python
numero = 0
```

```python
while(elelista < 100):
    numero = numero + 3
    lista.append(numero)
    elelista = len(lista)
```

### Toma el archivo datosCT.xlsx

1. Calcula la Utilidad (Ingreso - Gasto)
2. Encuentra el promedio de Gastos en los meses que la Utilidad fue negativa
3. Encuentra los ingresos totales del año

```python
excel2 = pd.read_excel("datosCT.xlsx")
```

```python
excel2
```

```python
excel2["Utilidad"] = excel2["Ingreso"] - excel2["Gasto"]
```

```python
excel2
```

```python
excel2["Gasto en mes con pérdidas"] = excel2["Gasto"].where(excel2["Utilidad"]<0)
```

```python
excel2
```

```python
excel2["Gasto en mes con pérdidas"].mean()
```

```python
IngresosTotales = excel2["Ingreso"].sum()
```
